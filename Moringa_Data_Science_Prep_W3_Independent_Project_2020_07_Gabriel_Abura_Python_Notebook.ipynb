```
# Importing pandas module in order to work with dataframes
import pandas as pd
```

```
# Then we have to load all our data sets
# we shall start with loading our cells_geo_description dataset 
# Creating DataFrame from the datasets
dt = "cells_geo_description.xlsx"
df = pd.read_excel(dt, sep=";")
df
```

```
# Secondly we load our cells_geo dataset
# Then we create a data frame from the dataset
dt = "cells_geo.csv"
df_geo = pd.read_csv(dt, sep=';', delimiter=None)
df_geo.head(10)
```

```
# Loading our CDR_description dataset 
# Creating DataFrame from the datasets.
#
dt = "CDR_description.xlsx"
df = pd.read_excel(dt, sep=";")
df
```

```
# Loading our Telcom_dataset dataset 
# Creating DataFrame from the datasets.
#
dt = "Telcom_dataset.csv"
df1 = pd.read_csv(dt, sep=",", delimiter=None)
print(df1)
```

```
# Loading our Telcom_dataset2 dataset 
# Creating DataFrame from the datasets.
#
dt = "Telcom_dataset2.csv"
df2 = pd.read_csv(dt, sep=",", delimiter=None)
df2.head(10)
```

```
# Loading our Telcom_Dataset3 dataset 
# Creating DataFrame from the dataset.
#
dt = "Telcom_dataset3.csv"
df3 = pd.read_csv(dt, sep=",", delimiter=None)
df3
```

```
# Next step is to merge the datasets whereby we create
# a list containing all 4 of the datasets
# # Dropping all the rows with NUll values from the dataset
# Getting the count of individual cities to know how many time each city appears in the column.
# 
#
frames = [df_geo, df1, df2, df3]
df_dataset = pd.concat(frames, axis=1)
x = df_dataset.dropna()
print(x)
```

```
# Getting the Insights for Analysis
# We are guided by the formulated questions
# Question 1
# Which were the most used cities for the 3 days
# Count the column VILLES which refers to cities

result = pd.value_counts(x['VILLES'])
city_users = result.head()
c_max = result.max()
print(f"The city with the most users appears: {c_max} times\n")
print(f"The Most used cities in 3 days: \n CITY       COUNT\n{city_users}")
```

```
# Question 2
# Which region is most used for the 3 days

region = pd.value_counts(x['REGION'])
print(f"\nThe most used region in 3 days are:\n")
print(f"Region               Count\n\
-------------------------\n{region.head()}")
```

```
# Question 3
# We have seen Cocody has the most users
# lets now check hoe many users call, text and 
# use data in cocody
voice = x[(x['VILLES'] == 'COCODY') & (x['PRODUTC'] == "Voice")]
data = x[(x['VILLES'] == 'COCODY') & (x['PRODUTC'] == "data")]
sms = x[(x['VILLES'] == 'COCODY') & (x['PRODUTC'] == "sms")]

voice_count = pd.value_counts(voice['PRODUTC'])
data_count = pd.value_counts(data['PRODUTC'])
sms_count = pd.value_counts(sms['PRODUTC'])

print("The Number of Users who Accessed MTN Services in \
\n\n       COCODY city\n-----------------------------\n\
        In 3 Days\n-----------------------------")
print(f"Total Voice users: {voice_count}\n\nTotal data users: {data_count}\n\nTotal sms users: {sms_count}\n\n")
```

```
# Then next we check the users behaviour
# in the second most used city which is Abobo
#using data in abobo
 
voice = x[(x['VILLES'] == 'ABOBO') & (x['PRODUTC'] == "Voice")]
data = x[(x['VILLES'] == 'ABOBO') & (x['PRODUTC'] == "data")]
sms = x[(x['VILLES'] == 'ABOBO') & (x['PRODUTC'] == "sms")]

voice_count = pd.value_counts(voice['PRODUTC'])
data_count = pd.value_counts(data['PRODUTC'])
sms_count = pd.value_counts(sms['PRODUTC'])

print("The Number of Users who Accessed MTN Services in \
\n       ABOBO city\n-----------------------------\n\
        In 3 Days\n-----------------------------")
print(f"Total Voice users: {voice_count}\n\nTotal data users: {data_count}\n\nTotal sms users: {sms_count}\n\n")
```

```
# Then finally we get the sum of billing price
# between these two cities where the product sells most
# We begin with Cocody

v_cocody = x[x['VILLES'] == 'COCODY']
sum_value = v_cocody['VALUE'].sum()
print(f"The sum of the Billing price per day in Cocody city is\n\
------------------------------------------------------\n{sum_value}")
```

```
# sum of billiing costs in Abobo

v_abobo = x[x['VILLES'] == 'ABOBO']
sum_value = v_abobo['VALUE'].sum()
print(f"The sum of the Billing price per day in Abobo city is\n\
------------------------------------------------------\n{sum_value}")
```
